```python
Q:
    Plot using sns.relplot, line plot, box plot on fmri data set
```

```python
import seaborn as sns #  Python data visualization library based on matplotlib for statistical graphics
import pandas as pd # data analysis and manipulation tool
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='signal',y='timepoint',hue='subject',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',hue='subject',kind='line',sort=True,data=fmri)
```

```python
sns.catplot(x='timepoint',y='signal',hue='subject',kind='boxen',data=fmri.sort_values('subject'))
```

```python
sns.catplot(x='timepoint',y='signal',hue='subject',kind='violin',data=fmri.sort_values('subject'))
```
