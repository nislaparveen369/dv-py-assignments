```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
y=x*x
plt.plot(x,y,'go--',linewidth=1,markersize=5)
plt.xlabel('root')
plt.ylabel('square')
plt.title('square of numbers')
```

```python
x=[1,2,3,4,5,6,7]
y=[160,150,140,145,175,165,180]
plt.plot(x,y,'ro--')
plt.xlabel('day')
plt.ylabel('sale of company')
plt.title('weekly sale')
```

```python
x=[1,2,3,4,5,6,7]
y=[160,150,140,145,175,165,180]
z=[70,90,160,150,140,145,175]
plt.plot(x,y,'ro--')
plt.xlabel('day')
plt.ylabel('sale of company1')
plt.plot(x,z,'go--')
plt.zlabel('sale of company2')
plt.title('weekly sale')
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,'go--')
plt.xlabel('a')
plt.ylabel('b')
#plt.show
```

```python
# 3 by 3 subplots
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
plt.subplot(3,3,1)
plt.plot(x,y1,y2,'r--',linewidth=2,markersize=12)
plt.subplot(3,3,2)
plt.plot(y3,y4,y5,'g*--',linewidth=2,markersize=5)

```
